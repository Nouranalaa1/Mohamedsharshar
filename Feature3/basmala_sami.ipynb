```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\norah\Downloads\titanic_data.csv")
df
```

```python
df.drop(df[df['fare']>140].index, axis = 0, inplace =True)
```

```python
sns.set_palette('Set3')
bp = sns.boxplot(data = df, x = "fare",color='skyblue')
plt.show()
sp = sns.scatterplot(data = df, x = "fare", y=df["survived"],color='skyblue')
plt.show()
```

```python
# Calculate the mean and standard deviation of the column
mean_col = df['fare'].mean()
std_col = df['fare'].std()
```

```python
# Calculate the upper and lower limits
upper_limit = mean_col + 3 * std_col
lower_limit = mean_col - 3 * std_col
```

```python
# Print the results
print("Upper limit: ",upper_limit)
print("Lower limit:Â ",lower_limit)
df_outliers = df.loc[(df['fare'] > upper_limit) | (df['fare'] < lower_limit)]
df_outliers
```
