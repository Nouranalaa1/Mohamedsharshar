```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\norah\Downloads\titanic_data.csv")
df
```

```python
for col in df.columns:
    if df[col].dtype != 'object':
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.set_title(f'Boxplot of {col}')
        bp = sns.boxplot(data = df, x = col,color='skyblue')
        plt.show()
```

```python
sns.set_palette('Set3')
bp = sns.boxplot(data = df, x = "fare",color='skyblue')
plt.show()
sp = sns.scatterplot(data = df, x = "fare", y=df["survived"],color='skyblue')
plt.show()
```
